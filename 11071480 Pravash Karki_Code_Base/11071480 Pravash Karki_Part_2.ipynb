### 2.2 Data Cleaning and Merging

### Creating function to find different district in  two dataframe to perform Merge 

```python
# Function to find out the different district name in two dataframes

def find_difference(list1, list2):
    list_difference = []  # stores the differences
    for item in list1:
        if item not in list2:
            list_difference.append(item)  #appends the different values found
    return list_difference
```

```python
"""Creating list of districts to use it in above function."""

horse_dis = list(df_horse.DISTRICT)  # df_horse converted in list to find out the list of districts
milk_dis = list(df_milk.DISTRICT)  # df_milk converted in list to find out the list of districts
```

```python
"""Using above function to find different value in both the list"""

find_difference(horse_dis, milk_dis)
```

###  Transforming the value in both df_horse and df_milk to make it equal for merge 

```python
"""Transforming above different district in to same spelling"""

df_milk['DISTRICT'] = df_milk['DISTRICT'].replace(['TERHATHUM'], 'TERATHUM')
df_milk['DISTRICT'] = df_milk['DISTRICT'].replace(['E. REGION'], 'E.REGION')
df_milk['DISTRICT'] = df_milk['DISTRICT'].replace(['C. REGION'], 'C.REGION')
df_milk['DISTRICT'] = df_milk['DISTRICT'].replace(['W. REGION'], 'W.REGION')
df_milk['DISTRICT'] = df_milk['DISTRICT'].replace(['MW. REGION'], 'MW.REGION')
df_milk['DISTRICT'] = df_milk['DISTRICT'].replace(['FW. REGION'], 'FW.REGION')
```

### Merging df_horse and df_milk and create new dataframe df_merge1

```python
df_merge1 = pd.merge(df_horse, df_milk, on="DISTRICT", how='outer')
df_merge1.head()
```

### Finding Difference between df_merge1 and df_meat before merging

```python
"""Using find_difference function to find difference between two dataframes"""

horse_dis = list(df_merge1['DISTRICT'])
milk_dis = list(df_meat['DISTRICT'])
find_difference(horse_dis, milk_dis)
```

### Renaming necessary district name 

```python
"""Renaming different district name to merge"""

df_meat['DISTRICT'] = df_meat['DISTRICT'].replace(['TERHATHUM'], 'TERATHUM')
df_meat['DISTRICT'] = df_meat['DISTRICT'].replace(['SANKHUWASABHA'], 'SANKHUWASHAVA')
```

## 3. Exploratory Data Analysis

### 3.1 Milk Production Analysis

```python
# Extracting and creating dataframe of five different regions
data_milk = df_finalmerge[df_finalmerge['DISTRICT'].str.startswith(("FW.", "MW.", "W.", "E.", "C."))]
data_milk
```

```python
# Setting index for all filtered data to perform analysis
set_index = data_milk.set_index("DISTRICT", drop=False)
set_index

# print(data_milk.iloc[:,0])
```

```python
# Bar plots to visualize the milk production in different mountain, hills, terai
size = (30, 25)
f, axes = plt.subplots(3, 2)
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
set_index.iloc[5:8, 2:7].plot.bar(figsize=size, ax=axes[0][0], cmap="coolwarm", title="E.MOUNTAIN, E.HILLS, E.TERAI",
                                  fontsize=14)
set_index.iloc[8:11, 2:7].plot.bar(figsize=size, ax=axes[0][1], cmap="coolwarm", title="C.MOUNTAIN, C.HILLS, C.TERAI",
                                   fontsize=14)
set_index.iloc[11:14, 2:7].plot.bar(figsize=size, ax=axes[1][0], cmap="coolwarm", title="W.MOUNTAIN, W.HILLS, W.TERAI",
                                    fontsize=14)
set_index.iloc[14:17, 2:7].plot.bar(figsize=size, ax=axes[1][1], cmap="coolwarm",
                                    title="MW.MOUNTAIN, MW.HILLS, MW.TERAI", fontsize=14)
set_index.iloc[17:20, 2:7].plot.bar(figsize=size, ax=axes[2][0], cmap="coolwarm",
                                    title="FW.MOUNTAIN, FW.HILLS, FW.TERAI", fontsize=14)
set_index.iloc[0:5, 2:7].plot.bar(figsize=size, ax=axes[2][1], cmap="coolwarm",
                                  title="E.REGION, C.REGION, W.REGION, MW.REGION, FW.REGION", fontsize=14)
plt.show()  # Display a figure
```

### 3.2 Duck Egg Analysis

```python
# Extracting and creating dataframe of five different regions
duck_numbers = df_finalmerge[df_finalmerge['DISTRICT'].str.startswith(("FW.", "MW.", "W.", "E.", "C."))]
duck_numbers
```

```python
#Total duck egg production in different hill, mountains, terai regions in Nepal
duck_egg = duck_numbers[['DISTRICT', 'LAYING DUCK', 'DUCK EGG']]

#Reshaping
duck_reshape = pd.melt(duck_egg, id_vars="DISTRICT", value_name="Laying Egg", var_name="Duck and Egg")
plt.figure(figsize=(30, 10))

#barplot to visualize the data
sns.barplot(x="DISTRICT", y="Laying Egg", data=duck_reshape, hue="Duck and Egg", palette="autumn")
plt.title('Total duck egg production in different hill, mountains, terai regions in Nepal', fontsize=16, loc='left')
plt.xlabel('Development Region')
plt.show()  # Display a figure
```

### 3.3 Meat Production Analysis

```python
meat_productions = df_finalmerge[df_finalmerge['DISTRICT'].str.endswith("REGION")]
meat_productions
```

```python
#Total meat production in different Development Region of Nepal
meat_prod = meat_productions[['DISTRICT', 'BUFF', 'MUTTON', 'CHEVON', 'CHICKEN', 'DUCK MEAT', 'TOTAL MEAT']]

#Reshaping
meat_reshape = pd.melt(meat_prod, id_vars="DISTRICT", value_name="Meat Production", var_name="Type of Meat")
plt.figure(figsize=(20, 5))

#barplot to visualize the data
sns.lineplot(x="DISTRICT", y="Meat Production", data=meat_reshape, hue="Type of Meat", palette="YlOrRd")
plt.title('Bar Diagram to show meat production in different region of Nepal', fontsize=16, loc='left')
plt.xlabel('Development Region')
plt.show()  # Display a figure
```

### 3.4 Wool Production Analysis

### 3.5 Cotton Production Analysis

```python
# Extracting cotton data of only three districts
cotton_productions = df_finalmerge[df_finalmerge['DISTRICT'].str.startswith(("DANG", "BANKE", "BARDIYA"))]
cotton_productions
```

```python
# Total cotton production in three district of Nepal
cotton_prod = cotton_productions[['DISTRICT', 'AREA (Ha.)', 'PROD. (Mt.)', 'YIELD Kg/Ha']]

# Reshaping
cotton_reshape = pd.melt(cotton_prod, id_vars="DISTRICT", value_name="Cotton Production", var_name="Cotton Type")
plt.figure(figsize=(20, 5))
# plot the point plot
sns.pointplot(x="DISTRICT", y="Cotton Production", data=cotton_reshape, hue="Cotton Type")
plt.title('Bar Diagram to show cotton production in different district of Nepal', fontsize=16, loc='left')
plt.xlabel('District')
plt.show()  # Display a figure
```

### 3.6 HorseAsses Analysis based on Regions

```python
# Extracting horseasses data of different regions 
horse_numbers = df_finalmerge[df_finalmerge['DISTRICT'].str.endswith(("REGION", "HILLS", "MOUNTAIN", "TERAI"))]
horse_numbers
```

```python
# Total horse production in different Development Region in Nepal
horse_num = horse_numbers[['DISTRICT', 'Horses_Asses']]

# Reshaping
horse_reshape = pd.melt(horse_num, id_vars="DISTRICT", value_name="Horse Population", var_name="Horse")
plt.figure(figsize=(25, 5))

# histplot to visualize the data
sns.pointplot(x="DISTRICT", y="Horse Population", data=horse_reshape, hue="Horse", palette="cubehelix_r")
plt.title('Total horse production in different Development Region in Nepal', fontsize=16, loc='left')
plt.xlabel('Development Region')
plt.show()  # Display a figure
```

### 3.7 Heatmap Of Total productions in different Districts

```python
# correlation between the different production items
correlation = df_finalmerge[['TOTAL MILK PRODUCED', 'TOTAL EGG', 'SHEEP WOOL PRODUCED', 'TOTAL MEAT']].corr()

plt.figure(figsize=(11, 9))

# plot the heatmap
sns.heatmap(correlation, annot=True, cmap="YlGnBu")
plt.title('Heatmap- Total Production of all Items ', fontsize=16, loc='left')
plt.xlabel('Production Items', fontsize=10)
plt.ylabel('Production Items', fontsize=10)

plt.show()  # Display a figure
```
